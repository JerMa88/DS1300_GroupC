```python
import pandas as pd
dialogue = pd.read_csv("../data/marvel-dialogue-nlp/data/mcu.csv")
```

```python
dialogue
```

```python
dialogue['line'] = dialogue['line'].apply(lambda x: " ".join(x.lower() for x in x.split()))
dialogue['line'].head()
```

```python
dialogue['line'] = dialogue['line'].str.replace('[^\w\s]','')
dialogue['line'].head()
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
dialogue['line'] = dialogue['line'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
dialogue['line'].head()
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from textblob import Word
dialogue['line'] = dialogue['line'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
dialogue['line'].head()
```

```python
freq = pd.Series(' '.join(dialogue['line']).split()).value_counts()[:10]
freq
```

```python
freq = pd.Series(' '.join(dialogue['line']).split()).value_counts()[-10:]
freq
```

```python
freq = list(freq.index)
dialogue['line'] = dialogue['line'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
dialogue['line'].head()
```

```python
import nltk
nltk.download('punkt')
```

```python
from textblob import TextBlob
f = []
for i in range(0,15723):
    f += TextBlob(dialogue['line'][i]).words
```

```python
f
```
